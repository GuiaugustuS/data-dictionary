```python
# Script desenvolvido para obter o template/layout e tipo de dados de cada fornecedor

import csv
from pyspark.sql import SparkSession

# Inicia uma sessão Spark
spark = SparkSession.builder.getOrCreate()

# Carrega o DataFrame original
df = spark.read.parquet(r"G:\.shortcut-targets-by-id\1kkcO2EWqo7VqwAVtJEGVmYuTF858en0s\HDI_DR\Arquivos_Source_Individuais\2023\Agger_Janeiro\calculos-auto-hdi-2023-01-01-2023-01-07.csv_parte_1.parquet", header=True, inferSchema=True)

# Obtém os tipos de dados de cada coluna
column_types = df.dtypes

# Dicionário que receberá os dados
template_padrao = {}

# Nome do arquivo CSV de saída
nome_arquivo = r'C:\Users\Guilherme\Documents\Talent Pool\Tropa BI\HDI\Multicalculo_validação\templates_padrao\template_padrao.csv'

# Abrir o arquivo CSV para escrita
with open(nome_arquivo, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Escrever o cabeçalho do CSV
    writer.writerow(['column_name','data_type'])

    # Percorrer as informações do dicionário
    for column in column_types:
        column_name = column[0]
        data_type = column[1]
        template_padrao = {column_name: data_type}
        
        # Escrever cada linha no arquivo CSV
        writer.writerow([column_name, data_type])


print(f"As informações foram salvas no arquivo {nome_arquivo}")
```
