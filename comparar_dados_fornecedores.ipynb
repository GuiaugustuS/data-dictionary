```python
import pandas as pd

path = r"C:\Users\Guilherme\Documents\Talent Pool\Tropa BI\HDI\Multicalculo_validação\Dicionário de dados\amostras.xlsx"

# Armazenando os dados de cada aba do excel em seu respectivo dataframe

df_segfy = pd.read_excel(path, sheet_name=0)

df_agger = pd.read_excel(path, sheet_name=1)

df_infocap = pd.read_excel(path, sheet_name=2)

df_quiver = pd.read_excel(path, sheet_name=3)

df_tex = pd.read_excel(path, sheet_name=4)

```

```python
list_df = [df_segfy, df_agger, df_infocap, df_quiver, df_tex]

# print(list_df)

for df in list_df:
    if df is df_segfy:
        novos_nomes = {}
        for column in df_segfy.columns:            
            novo_nome = f'SEGFY_{column}'
            novos_nomes[column] = novo_nome
        df_segfy = df.rename(columns = novos_nomes) 

    if df is df_agger:
        novos_nomes = {}
        for column in df_agger.columns:            
            novo_nome = f'AGGER_{column}'
            novos_nomes[column] = novo_nome
        df_agger = df.rename(columns = novos_nomes) 

    if df is df_infocap:
        novos_nomes = {}
        for column in df_infocap.columns:            
            novo_nome = f'INFOCAP_{column}'
            novos_nomes[column] = novo_nome
        df_infocap = df.rename(columns = novos_nomes) 

    if df is df_quiver:
        novos_nomes = {}
        for column in df_quiver.columns:            
            novo_nome = f'QUIVER_{column}'
            novos_nomes[column] = novo_nome
        df_quiver = df.rename(columns = novos_nomes) 

    if df is df_tex:
        novos_nomes = {}
        for column in df_tex.columns:            
            novo_nome = f'TEX_{column}'
            novos_nomes[column] = novo_nome
        df_tex = df.rename(columns = novos_nomes)
```

```python
df_final = pd.concat([df_segfy, df_agger, df_infocap, df_quiver, df_tex], ignore_index=False, axis=1)

df_final
```

```python
df_final.to_csv(r'C:\Users\Guilherme\Documents\Talent Pool\Tropa BI\HDI\Multicalculo_validação\Dicionário de dados\amostra_total.csv',index=False, sep=',', encoding='utf-8',header=True)
```
